```python
res = qtnacs.time_indep_unitary(H, [1], 1/np.sqrt(2), 100, [[0, 0], [0,1],[1,0], [1,1]], True)
```

```python
def driven_ryd(omega, det, coords, interaction_coeff, atom_types, powers):
    def cb(init_state):
        H = qtnacs.DrivenRyd(omega, det, coords, interaction_coeff, atom_types, powers)
        res = qtnacs.time_indep_unitary(H, init_state, 1/(omega / (2 * np.pi) * np.sqrt(2)), 100, [], False)
        return res.final_state
    return cb
```

```python
fid, op = qtnacs.get_fidelity(driven_ryd(omega, det, coords, interaction_coeff, atom_types, powers), qtnacs.cGate([1,2], qt.sigmaz()))
```
